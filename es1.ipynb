# Esercizio 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint


#Let's call phi_dot the prime derivative of phi.


def function_ode(pos, time):
    phi, phi_dot = pos

    ode1 = phi_dot
    ode2 = -omega**2*phi-lam*phi**3

    return np.array([ode1,ode2])


lam = 0
omega = 1

time = np.linspace(0,50,10000)
pos_iniziali = np.array([2,0])

sol = odeint(function_ode, pos_iniziali, time)

# solve with lambda=0.1

lam = 0.1
sol2 = odeint(function_ode, pos_iniziali, time)


plt.plot(time, sol[:,0],label="lambda = 0")
plt.plot(time, sol2[:,0], label="lambda=0.1")
plt.legend()

```

## q 1.2

```python
plt.plot(sol[:,0],sol[:,1],label="lambda = 0")
plt.plot(sol2[:,0],sol2[:,1],label="lambda = 0.1")

plt.xlabel("phi")
```

## q1.3

```python
def energy(phi,phi_dot):

    return 0.5*phi_dot**2+0.5*omega**2*phi**2 + lam/4 * phi**4


energy_over_time = energy(sol2[:,0],sol2[:,1])

plt.plot(time, energy_over_time)
#look closely at the scale


max_discrepancy = np.max(np.abs(energy_over_time-energy_over_time[0])/energy_over_time[0])


print(f"max discrepancy is {max_discrepancy*100} %")
```
